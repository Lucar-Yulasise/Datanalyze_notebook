```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
an = plt.imread("./cv2_change_head/安倍.jpg")
```

```python
an.shape
```

```python
plt.imshow(an)
```

```python
# 导入cv2库，识别图像某些特征的库，也叫做计算机视觉库
import cv2

# 用cv2这个视觉库，创建算法模型
cade = cv2.CascadeClassifier() # 这个模型用于加载一些第三方的视觉算法
# 加载人脸识别的算法
cade.load("./cv2_change_head/haarcascade_frontalface_default.xml")
#现在cade这个模型已经具备识别人脸的能力
```

```python
# 用视觉模型来识别安倍的脸的范围
faces = cade.detectMultiScale(an) 
```

```python
faces
```

```python
girls = plt.imread("./cv2_change_head/girls.jpg")
g_faces = cade.detectMultiScale(girls)
```

```python
plt.imshow(girls)
```

```python
g_faces
```

```python
dog = plt.imread("./cv2_change_head/dog.jpg")
d_faces = cade.detectMultiScale(dog)
```
