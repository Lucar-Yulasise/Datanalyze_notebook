根据日期对全体数据进行排序

 把日期标准化



```python
def convert_dt(date):
    day,month,year = date.split("-")
    m = months[month]
    return "20" + str(year) + "-" + str(m) +"-" + str(day)
# 1-jun-11 => 2011-06-01
```

```python
ele["contb_receipt_dt"] = ele["contb_receipt_dt"].map(convert_dt)
```

```python
# 日期现在还是字符串，需要转成日期
ele["contb_receipt_dt"] = pd.to_datetime(ele["contb_receipt_dt"])
ele.dtypes
```
