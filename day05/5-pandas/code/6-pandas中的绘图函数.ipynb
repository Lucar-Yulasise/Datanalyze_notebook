Series和DataFrame都有一个用于生成各类图表的plot方法。默认情况下，它们所生成的是线形图

```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame

%matplotlib inline
```

## 线形图

```python
s = Series([90,12,1,0,45,23,89],name="python",index=list("abcdefg"))
s.plot()
# index为x轴坐标，values为y轴坐标
```

```python
df = DataFrame(np.random.randint(0,150,size=(4,4)),
               index=["张老三","Jack","Rose","Laowang"],
               columns=["数学","python","java","English"])
df
```

```python
df.plot() 
# 【注意】由于matplotlib不支持中文的显示
# index是x轴坐标，values通过多个图像来表示y轴坐标，columns用不同的图像和对应的图例来表示
```

## 柱状图

```python
s.plot(kind="bar",colormap="gray")
```

```python
df.plot(kind="bar")
```

## 散布图

```python
x = np.linspace(-10,10,50)
y = x**2
```

```python
df = DataFrame({
    "A":x,
    "B":y
})
df
```

```python
df.plot(kind="scatter",x="A",y="B")
```

##### 随机数百分比密度图

```python
n1 = np.random.normal(0,1,size=200)
n2 = np.random.normal(10,4,size=200)
n = np.concatenate([n1,n2])
s = Series(n) 
```

```python
s.hist(bins=50,normed=True)
s.plot(kind="kde") # 反映的统计学规律类似于直方图
```
